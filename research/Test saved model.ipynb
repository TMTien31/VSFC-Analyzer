```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer_sentiment = AutoTokenizer.from_pretrained("tmt3103/VSFC-sentiment-classify-phoBERT")
model_sentiment = AutoModelForSequenceClassification.from_pretrained("tmt3103/VSFC-sentiment-classify-phoBERT")

tokenizer_topic = AutoTokenizer.from_pretrained("tmt3103/VSFC-topic-classify-phoBERT")
model_topic = AutoModelForSequenceClassification.from_pretrained("tmt3103/VSFC-topic-classify-phoBERT")
```

```python
import torch
```

1. Chuyển mô hình sang chế độ đánh giá (xem trong file train sentiment)
2. torch.no_grad() để tắt chế độ theo dõi gradient
3. encoding input
4. đưa vào model
5. tìm giá trị lớn nhất theo hàng ngang, trả về index của giá trị đó, cũng là nhãn dự đoán

```python
def predict_sentiment(text):
    model_sentiment.eval()
    with torch.no_grad():
        encoding = tokenizer_sentiment(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
        output = model_sentiment(**encoding)
        pred = torch.argmax(output.logits, dim=1).item()
    return pred  # sẽ là 0, 1, hoặc 2
```

```python
def predict_topic(text):
    model_topic.eval()
    with torch.no_grad():
        encoding = tokenizer_topic(text, return_tensors='pt', truncation=True, padding=True, max_length=128)
        output = model_topic(**encoding)
        pred = torch.argmax(output.logits, dim=1).item()
    return pred  # sẽ là 0, 1, 2 hoặc 3
```

```python
id2label_sentiment = {0: "tiêu cực", 1: "trung lập", 2: "tích cực"}
id2label_topic = {0: "giảng viên", 1: "chương trình đào tạo", 2: "cơ sở vật chất", 3: "vấn đề khác"}

input = "Giảng viên take care tốt ác" # Nhập ở đây(feedback)
label_sentiment = id2label_sentiment[predict_sentiment(input)]
label_topic = id2label_topic[predict_topic(input)]

print("Feedback:", input)
print(f"Đây là feedback {label_sentiment} về {label_topic}")
```
